## 1) Propuesta de valor
- Resolver problemas de clasificacion, regresion y analisis estadistico.
- Entregar modelos con validacion y metricas claras.
- Comunicar resultados de forma ejecutiva.

```python
from pathlib import Path
import pandas as pd


def get_repo_root(start: Path | None = None) -> Path:
    base = (start or Path.cwd()).resolve()
    for candidate in [base, *base.parents]:
        if (candidate / "README.md").exists() and (candidate / "scripts").exists():
            return candidate
    return base


ROOT = get_repo_root()
print(f"Repo root: {ROOT}")

files = [
    ROOT / "reports/iris_model_ranking.csv",
    ROOT / "reports/titanic_openml_model_ranking.csv",
    ROOT / "reports/wine_quality_model_ranking.csv",
]
for f in files:
    print(f.relative_to(ROOT), "OK" if f.exists() else "MISSING")
```

## 2) Ranking de modelos
Comparativa directa para elegir baseline competitivo.

```python
iris = pd.read_csv(ROOT / "reports/iris_model_ranking.csv")
titanic = pd.read_csv(ROOT / "reports/titanic_openml_model_ranking.csv")
wine = pd.read_csv(ROOT / "reports/wine_quality_model_ranking.csv")

print("Iris")
display(iris)
print("\nTitanic/Fallback")
display(titanic)
print("\nWine/Fallback")
display(wine)
```
